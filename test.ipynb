```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("bookstore.db")

# Load books
books = pd.read_sql("SELECT * FROM books", conn)

# Book price statistics
mean_price = books["price"].mean()
median_price = books["price"].median()
min_price = books["price"].min()
max_price = books["price"].max()
price_range = max_price - min_price
variance_price = books["price"].var()
std_price = books["price"].std()

print("Mean book price:", mean_price)
print("Median book price:", median_price)
print("Minimum book price:", min_price)
print("Maximum book price:", max_price)
print("Range of book prices:", price_range)
print("Variance of book prices:", variance_price)
print("Standard deviation of book prices:", std_price)

# Join orders and books
sales_df = pd.read_sql("""
SELECT 
    o.order_id,
    o.customer_id,
    o.book_id,
    o.quantity,
    o.order_date,
    b.title,
    b.genre,
    b.price,
    (o.quantity * b.price) AS sale_amount
FROM orders o
JOIN books b
ON o.book_id = b.book_id
""", conn)

print("\nSales Data:")
print(sales_df)

# Total sales
total_sales = sales_df["sale_amount"].sum()
print("\nTotal sales:", total_sales)

# Average order value
average_order_value = sales_df["sale_amount"].mean()
print("Average order value:", average_order_value)

# Most purchased category
genre_sales = sales_df.groupby("genre")["quantity"].sum().sort_values(ascending=False)
print("\nQuantity sold by genre:")
print(genre_sales)

most_purchased_category = genre_sales.idxmax()
print("\nMost purchased category:", most_purchased_category)
```
