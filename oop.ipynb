```python
class Book():
    def __init__ (self,title,price,stock,category,is_available=True):
        self.title = title
        self.price = price
        self.stock = stock
        self.category = category
        self.is_available = is_available
    
    def display_info(self):
        print(f"Title:{self.title}")
        print(f"Price: {self.price}")
        print(f"Stock: {self.stock}")
        print(f"category {self.category}")

    def update_stock(self, amount):
      self.stock += amount
      self.is_available = self.stock > 0
    
    def checkout(self,quantity=1):
        if self.stock >= quantity:
            self.stock -= quantity
            if self.stock == 0:
                self.is_available = False
        else:
            print("Not enough stock available.")
        
    def return_item(self,quantity=1):
        self.stock += quantity
        self.is_available = True
        

class Customer:
    def __init__(self,customer_name, email, city):
        self.customer_name = customer_name
        self.email = email
        self.city = city
    
    def place_order(self):
        print(f"Customer_name: {self.customer_name}")
        print(f"Email: {self.email}")
        print(f"City: {self.city}")
    
    def display_customer(self):
        print(f"Customer_name: {self.customer_name}")
        print(f"Email: {self.email}")
        print(f"City: {self.city}")


class Order:
    def __init__(self, order_id, customer,items):
        self.order_id = order_id
        self.customer = customer
        self.items = items
    
    def calculate_total(self):
      total = 0
      for book, quantity in self.items:
          total += book.price * quantity
      return total

    def add_item(self, book, quantity):
      if book.stock >= quantity:
          self.items.append((book,quantity))
          book.checkout(quantity)
          print(f"Added {quantity} x {book.title} to order.")
      else:
          print(f"Cannot add {book.title} Not enough stock.")

    def print_receipt(self):
        print(f"\nOrder ID: {self.order_id}")
        print(f"Customer: {self.customer.customer_name}")
        print("items")
        for book, quantity in self.items:
            print(f"- {book.title} x {quantity} = {book.price * quantity}")
        print(f"Total: {self.calculate_total}")

```

```python
# Create a book
book = Book("With no Wifi", 14900, 19, "Reality")
book.display_info()

# Update stock
book.update_stock(5)
print("\nAfter adding stock:")
book.display_info()

# Checkout one copy
book.checkout()
print("\nAfter checkout:")
book.display_info()

# Return one copy
book.return_item()
print("\nAfter return:")
book.display_info()

print("\n1\n")

# Create customer
customer = Customer("Drey WIFI", "savagebat23@gmail.com", "Miami")
customer.display_customer()

print("\n2\n")

# Create order
order = Order(13618, customer)

# Add book to order
order.add_item(book, 2)

# Customer places order
customer.place_order(order)

# Print receipt
order.print_receipt()
```
